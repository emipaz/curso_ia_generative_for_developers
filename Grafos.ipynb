# ¿Qué es un grafo?

GPT : Un grafo es una estructura de datos fundamental que consiste en un conjunto de nodos (o vértices) conectados por aristas (o arcos). Los grafos se utilizan para representar relaciones entre entidades y pueden ser útiles en varios campos, como la informática, la matemática, la biología y la ingeniería.

**Elementos del grafo:**

1. **Vértices**: Son los puntos o nodos del grafo. Representan las entidades o elementos que se están modelando.
  
2. **Aristas**: Son las conexiones o relaciones entre los vértices. Pueden ser dirigidas (si tienen una dirección) o no dirigidas (si la conexión es bidireccional).

## Tipos de grafos:

1. **Grafo dirigido (digrafo)**: Las aristas tienen una dirección, lo que significa que la relación de un vértice a otro no es necesariamente recíproca.

2. **Grafo no dirigido**: Las aristas no tienen dirección, lo que permite que la relación entre dos vértices sea bidireccional.

3. **Grafo ponderado**: Las aristas tienen un peso o costo asociado, lo que puede significar la distancia, tiempo, costo, etc., entre los vértices.

4. **Grafo conexo**: Hay un camino (una secuencia de aristas) entre cualquier par de vértices.

5. **Grafo completo**: Cada par de vértices está conectado por una arista.

### Implementación de un Grafo en Python

````python
class Grafo:
    def __init__(self):
        # Inicializa un Grafo como un diccionario vacio
        self.adjacencias = {}

    def agregar_vertice(self, vertice):
        # Adds a vertex to the graph
        if vertice not in self.adjacencias:
            self.adjacencias[vertice] = []

    def agregar_arista(self, vertice1, vertice2, dirigido=False):
        # Adds an edge between two vertices
        if vertice1 not in self.adjacencias:
            self.agregar_vertice(vertice1)
        if vertice2 not in self.adjacencias:
            self.agregar_vertice(vertice2)

        self.adjacencias[vertice1].append(vertice2)
        if not dirigido:
            self.adjacencias[vertice2].append(vertice1)

    def obtener_vertices(self):
        # Returns all vertices in the graph
        return list(self.adjacencias.keys())

    def obtener_aristas(self):
        # Returns all edges in the graph
        aristas = []
        for vertice, vecinos in self.adjacencias.items():
            for vecino in vecinos:
                aristas.append((vertice, vecino))
        return aristas

    def __str__(self):
        # String representation of the graph
        resultado = ""
        for vertice, vecinos in self.adjacencias.items():
            resultado += f"{vertice}: {', '.join(map(str, vecinos))}\n"
        return resultado.strip()
````

````python
import threading

class Graph:
    def __init__(self, directed=False):
        self.graph = {}
        self.directed = directed
        self.lock = threading.Lock()  # Para manejo de concurrencia
    
    def add_vertex(self, vertex):
        with self.lock:  # Uso de bloqueos para evitar condiciones de carrera
            if not isinstance(vertex, (int, str, tuple)):
                raise ValueError("Vertex must be a hashable type.")
            if vertex not in self.graph:
                self.graph[vertex] = set()  # Cambiado a set para mejor eficiencia
    
    def add_edge(self, src, dest, weight):
        with self.lock:
            if src not in self.graph or dest not in self.graph:
                raise KeyError("Both vertices must exist in the graph.")
            if dest not in self.graph[src]:  # Check to prevent duplicate edges
                self.graph[src][dest] = weight
            if not self.directed and src not in self.graph[dest]:
                self.graph[dest][src] = weight

    def remove_edge(self, src, dest):
        with self.lock:
            self.graph.get(src, {}).pop(dest, None)  # Elimina sin lanzar excepción si no existe
            if not self.directed:
                self.graph.get(dest, {}).pop(src, None)
    
    def remove_vertex(self, vertex):
        with self.lock:
            if vertex in self.graph:
                for adj in list(self.graph.keys()):
                    self.graph[adj].discard(vertex)  # Evita errores si no existe el vértice
                del self.graph[vertex]
    
    def get_adjacent_vertices(self, vertex):
        with self.lock:
            return list(self.graph.get(vertex, set()))  # Devuelve una lista de adyacentes
    
    def _get_edge_weight(self, src, dest):
        with self.lock:
            return self.graph.get(src, {}).get(dest, float('inf'))
    
    def __str__(self):
        with self.lock:
            return str({k: list(v) for k, v in self.graph.items()})  # Convertir sets a listas para mejor lectura
````

- **Uso de `set`**: Los sets son perfectos para representar listas de vecinos en un grafo porque no permiten duplicados y tienen un tiempo de búsqueda amortiguado O(1), lo que mejora la eficiencia al eliminar comprobaciones duplicadas.

- **Manejo de excepciones**: Usar `pop()` con un valor predeterminado en lugar de eliminar con `del`, previene errores si se intenta eliminar un vértice o arista no existente.

- **Bloqueos y Concurrencia**: Al envolver operaciones críticas con un contexto `with self.lock`, se previenen problemas en aplicaciones potencialmente multihilo, donde múltiples hilos intentan modificar la estructura de datos al mismo tiempo.

````python
import threading

class Graph:
    def __init__(self, directed=False):
        """
        Inicializa el grafo.

        Parámetros:
        - directed (bool): Especifica si el grafo es dirigido. 
          El valor por defecto es False (no dirigido).
        
        Atributos:
        - graph (dict): Un diccionario para almacenar los vértices 
          y sus vértices adyacentes junto con sus pesos.
        - directed (bool): Indica si el grafo es dirigido.
        - lock (threading.Lock): Un objeto de bloqueo para manejar la concurrencia.
        """
        self.graph = {}  # Almacena los vértices y un dict para sus vecinos y pesos
        self.directed = directed
        self.lock = threading.Lock()  # Para manejo de concurrencia
    
    def add_vertex(self, vertex):
        """
        Agrega un vértice al grafo.

        Parámetros:
        - vertex: El vértice a agregar, debe ser un tipo hashable.

        Lanza:
        - ValueError: Si el vértice no es un tipo hashable.
        """
        with self.lock:  # Uso de bloqueos para evitar condiciones de carrera
            if not isinstance(vertex, (int, str, tuple)):
                raise ValueError("El vértice debe ser un tipo hashable.")
            if vertex not in self.graph:
                self.graph[vertex] = {}  # Mantiene un dict para adyacentes y pesos
    
    def add_edge(self, src, dest, weight):
        """
        Agrega una arista ponderada de src a dest. 
        Si el grafo es no dirigido, también agrega de dest a src.

        Parámetros:
        - src: El vértice fuente.
        - dest: El vértice destino.
        - weight: El peso de la arista.

        Lanza:
        - KeyError: Si alguno de los vértices no existe en el grafo.
        """
        with self.lock:
            if src not in self.graph or dest not in self.graph:
                raise KeyError("Ambos vértices deben existir en el grafo.")
            if dest not in self.graph[src]:  # Previene aristas duplicadas
                self.graph[src][dest] = weight  # Guarda el peso de la arista
            if not self.directed and src not in self.graph[dest]:
                self.graph[dest][src] = weight  # Para grafos no dirigidos
    
    def remove_edge(self, src, dest):
        """
        Elimina una arista de src a dest. 
        Si el grafo es no dirigido, también elimina de dest a src.

        Parámetros:
        - src: El vértice fuente.
        - dest: El vértice destino.
        """
        with self.lock:
            self.graph.get(src, {}).pop(dest, None)  # Elimina sin lanzar excepción si no existe
            if not self.directed:
                self.graph.get(dest, {}).pop(src, None)  # Elimina arista de vuelta si no es dirigido

    def remove_vertex(self, vertex):
        """
        Elimina un vértice y todas las aristas conectadas a él.

        Parámetros:
        - vertex: El vértice que se va a eliminar.
        """
        with self.lock:
            if vertex in self.graph:
                # Remueve aristas conectadas a este vértice de otros vértices
                for adj in list(self.graph.keys()):
                    self.graph[adj].pop(vertex, None)  # Elimina sin lanzar excepción
                del self.graph[vertex]  # Elimina el vértice

    def get_adjacent_vertices(self, vertex):
        """
        Obtiene una lista de vértices adyacentes al vértice especificado.

        Parámetros:
        - vertex: El vértice cuyos vecinos se van a recuperar.

        Devuelve:
        - Lista de vértices adyacentes. Devuelve una lista vacía si el vértice no se encuentra.
        """
        with self.lock:
            return list(self.graph.get(vertex, {}).keys())  # Devuelve los vértices adyacentes
    
    def _get_edge_weight(self, src, dest):
        """
        Obtiene el peso de la arista de src a dest.

        Parámetros:
        - src: El vértice fuente.
        - dest: El vértice destino.

        Devuelve:
        - El peso de la arista. Si la arista no existe, devuelve infinito.
        """
        with self.lock:
            return self.graph.get(src, {}).get(dest, float('inf'))  # Devuelve el peso o infinito
    
    def __str__(self):
        """
        Proporciona una representación en cadena de la lista de adyacencia del grafo.
        
        Devuelve:
        - Una representación en cadena del diccionario del grafo.
        """
        with self.lock:
            return str(self.graph)  # Muestra la representación en cadena actual del grafo
````

Para implementar el método `shortest_path` en la clase `Graph_Advanced`, podemos utilizar el algoritmo de Dijkstra, que es eficiente para encontrar el camino más corto en un grafo ponderado. Este algoritmo funciona bien en grafos dispersos y nos permitirá cumplir el requerimiento de completar la búsqueda en menos de 0.5 segundos, siempre que el grafo no sea excesivamente grande o denso.

Aquí está la implementación del método `shortest_path`, incluyendo el uso de una cola de prioridad para gestionar los nodos a explorar:

````python
import heapq  # Importar heapq para usar como una cola de prioridad

class GrafoAvanzado(GrafoAvanzado):
    # Otros métodos de la clase...

    def shortest_path(self, start, end):
        """
        Calcular el camino más corto desde un nodo de inicio hasta un nodo de destino en un grafo disperso con potencialmente miles de nodos. 
        Debe ejecutarse en menos de 0.5 segundos y encontrar la distancia más corta entre dos nodos.

        Parámetros:
        - start: El nodo de inicio.
        - end: El nodo de destino.

        Devuelve:
        Una tupla que contiene la distancia total del camino más corto y una lista de nodos que representa ese camino.
        """
        
        # Inicializar las estructuras de datos
        min_heap = []  # Cola de prioridad (min-heap)
        distances = {vertex: float('inf') for vertex in self.graph}  # Distancias iniciales a infinito
        previous_nodes = {vertex: None for vertex in self.graph}  # Almacena el nodo previo en el camino más corto

        # La distancia inicial desde el nodo de inicio es 0
        distances[start] = 0
        # Agregar el nodo de inicio a la cola de prioridad con distancia 0
        heapq.heappush(min_heap, (0, start))

        while min_heap:
            current_distance, current_vertex = heapq.heappop(min_heap)  # Obtener el nodo con la menor distancia
            
            # Si encontramos el nodo de destino, arreglar el camino y devolver resultados
            if current_vertex == end:
                path = []
                while current_vertex is not None:
                    path.append(current_vertex)
                    current_vertex = previous_nodes[current_vertex]
                path.reverse()  # Revertimos el camino para que vaya de inicio a final
                return distances[end], path

            # Si la distancia actual es mayor que la registrada, no hacer nada
            if current_distance > distances[current_vertex]:
                continue
            
            # Procesa todos los vecinos del nodo actual
            for neighbor in self.get_adjacent_vertices(current_vertex):
                weight = self._get_edge_weight(current_vertex, neighbor)  # Obtener el peso de la arista
                distance = current_distance + weight  # Calcular la nueva distancia

                # Solo considerar este nuevo camino si es mejor
                if distance < distances[neighbor]:
                    distances[neighbor] = distance  # Actualizar la distancia más corta
                    previous_nodes[neighbor] = current_vertex  # Actualizar el nodo previo
                    heapq.heappush(min_heap, (distance, neighbor))  # Agregarlo a la cola de prioridad

        return float('inf'), []  # Si no hay camino, devolver infinito y una lista vacía

    def shortest_path_opc2(self, start, end): 
        """
        Calcular el camino más corto desde un nodo de inicio hasta un nodo de destino en un grafo disperso con potencialmente miles de nodos. 
        Debe ejecutarse en menos de 0.5 segundos y encontrar la distancia más corta entre dos nodos.

        Parámetros:
        - start: El nodo de inicio.
        - end: El nodo de destino.

        Devuelve:
        Una tupla que contiene la distancia total del camino más corto y una lista de nodos que representa ese camino.
        """

        # creamos un diccionadio con las distancias en infinito
        distances = {vertex: float('inf') for vertex in self.graph}
                
        distances[start] = 0
        
        # Cola de prioridad para contener los vértices que se explorarán, inicializada con el nodo de inicio
        priority_queue = [(0, start)]
        
        # Diccionario para almacenar el camino recorrido para llegar a cada vértice
        previous_nodes = {vertex: None for vertex in self.graph}
        
        while priority_queue:
            # Obtenga el vértice con la menor distancia
            current_distance, current_vertex = heapq.heappop(priority_queue)
            
            # Si hemos llegado al nodo final, reconstruimos la ruta.
            if current_vertex == end:
                # creamos una ruta vacia
                path = []
                # mientesa el vertice no see nulo agregamos el vertice en la ruta
                while current_vertex is not None:
                    path.append(current_vertex) # agregamos el vertice actual en la ruta vamos del final al inicio
                    current_vertex = previous_nodes[current_vertex] # cambiamos el vertice_Actual y seguir el while
                # al terminar el while debolvemos la ruta invertida desde il inicio al final    
                return current_distance, path[::-1]
            
            # Si la distancia actual es mayor que la distancia más corta registrada, omitir
            if current_distance > distances[current_vertex]:
                continue
            
            # Explorar vecinos
            for neighbor, weight in self.graph[current_vertex].items():
                distance = current_distance + weight
                
                # Si se encuentra un camino más corto hacia el vecino
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    previous_nodes[neighbor] = current_vertex # guardamos el vecino con menor distancia en previus_nodes
                    # añadir el vecino a la cola de prioridad (`priority_queue`), junto con su distancia
                    heapq.heappush(priority_queue, (distance, neighbor)) 
        
        # Si el nodo final es inalcanzable, devuelve infinito y una ruta vacía
        return float('inf'), []
        # return dist, path
````

### Explicación del Método `shortest_path`:

1. **Inicialización**:
   - Se establece una cola de prioridad (`min_heap`) para gestionar los nodos que se explorarán.
   - Se inicializan diccionarios para almacenar las distancias mínimas desde el nodo de inicio a cada nodo (`distances`) y el nodo anterior en el camino más corto (`previous_nodes`).
2. **Distancias Iniciales**:
   - Se establece la distancia del nodo de inicio a sí mismo como 0 y a todos los demás nodos como infinito.
3. **Exploración**:
   - En un bucle `while`, se extrae el nodo con la menor distancia de la cola de prioridad.
   - Si se llega al nodo de destino, se recupera el camino y se devuelve la distancia y el camino.
   - Si la distancia actual es mayor que la distancia registrada en `distances`, se ignora (porque ya hemos encontrado un camino más corto anteriormente).
4. **Procesamiento de Vecinos**:
   - Para cada vecino del nodo actual, se calcula la distancia acumulada.
   - Si esta nueva distancia es la más corta conocida hasta el vecino, se actualizan las distancias y la cola de prioridad.
5. **Sin Camino Encontrado**:
   - Si se agotan los nodos en la cola sin haber alcanzado el nodo de destino, se devuelve infinito y una lista vacía.

Para resolver el Problema del Viajante (TSP, por sus siglas en inglés) para un grafo completo con un máximo de 10 nodos, podemos utilizar un enfoque de **fuerza bruta** mediante una búsqueda de todas las permutaciones de los nodos. Esto es viable debido a que el número de permutaciones para un grafo de hasta 10 nodos es $10! (3,628,800)$, lo cual es computacionalmente manejable en menos de 1 segundo en un entorno moderno.

A continuación se encuentra la implementación del método `tsp_small_graph`, que calcula la ruta más corta comenzando desde un nodo especificado:

````python
import itertools  # Importar itertools para generar permutaciones

class GrafoAvanzado(GrafoAvanzado):
    # Otros métodos de la clase...

    def tsp_small_graph(self, start_vertex):
        """
        Resuelve el Problema del Viajante de Comercio para un grafo completo pequeño (~10 nodos) comenzando desde un nodo especificado. 
        Se requiere encontrar el recorrido óptimo. Se esperan grafos con un máximo de 10 nodos. Debe ejecutarse en menos de 1 segundo.

        Parámetros:
        - `start_vertex`: El nodo de inicio.

        Retornos:
        - Una tupla que contiene la distancia total del recorrido y una lista de nodos que representan el camino del recorrido.
        """

        # Obtener todos los nodos excepto el nodo de inicio
        nodes = self.get_adjacent_vertices(start_vertex)
        n = len(nodes)

        # Inicializar variables para el mejor costo y el mejor camino
        best_cost = float('inf')
        best_path = []

        # Generar todas las permutaciones de los nodos
        for perm in itertools.permutations(nodes):
            # Calcular la distancia total de esta permutación iniciando desde start_vertex
            current_cost = 0
            current_path = [start_vertex] + list(perm)  # Empezar el camino

            # Calcular la distancia de la ruta
            for i in range(len(current_path)):
                if i == len(current_path) - 1:  # Regresar al nodo de inicio
                    current_cost += self._get_edge_weight(current_path[i], start_vertex)
                else:
                    current_cost += self._get_edge_weight(current_path[i], current_path[i + 1])

            # Si la ruta actual es mejor, actualizar mejor costo y mejor camino
            if current_cost < best_cost:
                best_cost = current_cost
                best_path = current_path
                
        return best_cost, best_path  # Retorna el mejor costo y el mejor camino
````

### Explicación del Método `tsp_small_graph`:

1. **Obtención de Nodos**:

2. **Inicialización**:

3. **Generación de Permutaciones**:
   - Se utilizan las combinaciones de `itertools.permutations` para generar todas las posibles rutas que incluyen todos los nodos adyacentes al nodo de inicio.
4. **Cálculo de Distancia para cada Permutación**:
   - Para cada permutación, se calcula el costo total del recorrido:
     - Se suman los pesos de las aristas correspondientes a cada paso en la permutación, incluyendo el regreso al nodo de inicio al final.
5. **Actualización del Mejor Camino**:
   - Si el costo de la ruta actual es menor que el mejor costo encontrado, se actualizan tanto el mejor costo como el mejor camino.
6. **Devolución de Resultados**:
   - Al final, se retorna el mejor costo encontrado y el camino que produce esa distancia.

## El método por fuerza bruta no cumple con el requerimiento de tiempo necesario, busca un método mas eficiente

Para resolver el Problema del Viajante (TSP) de manera más eficiente, especialmente en grafos pequeños (hasta 10 nodos), podemos utilizar una técnica conocida como **programación dinámica**. Este método, conocido como el algoritmo de Held-Karp, reduce significativamente la cantidad de cálculos necesarios en comparación con la búsqueda de permutaciones.

La idea básica de este enfoque es almacenar soluciones parciales en una tabla para que no tengamos que recalcular el costo de caminos que ya hemos computado. Este método tiene una complejidad de tiempo de $O(n^2 \cdot 2^n)$, lo que es manejable para 10 nodos.

Aquí te muestro cómo se puede implementar este enfoque:

````python
class GrafoAvanzado(GrafoAvanzado):
    # Otros métodos de la clase...

    def tsp_small_graph_opc2(self, start_vertex):
        """
        Resuelve el Problema del Viajante de Comercio para un grafo completo pequeño (~10 nodos) comenzando desde un nodo especificado.
        Se requiere encontrar el recorrido óptimo. Se esperan grafos con un máximo de 10 nodos. Debe ejecutarse en menos de 1 segundo.

        Parámetros:
        - `start_vertex`: El nodo de inicio.

        Retornos:
        - Una tupla que contiene la distancia total del recorrido y una lista de nodos que representan el camino del recorrido.
        """
        # Obtener todos los nodos adyacentes en el grafo
        nodes = self.get_adjacent_vertices(start_vertex)
        n = len(nodes)

        # Preparamos la estructura de memoria para los costos
        memo = {}
        # Guardamos la ruta para recrearla más tarde
        path = {}

        # Función recursiva para calcular el costo usando cache
        def tsp(visited, current_node):
            # Si todos los nodos han sido visitados, volver al inicio
            if len(visited) == n:
                return self._get_edge_weight(current_node, start_vertex)  # Regresar al nodo de inicio

            # Usar una clave única para el estado actual
            key = (tuple(visited), current_node)
            if key in memo:
                return memo[key]

            min_cost = float('inf')

            # Explorar los nodos adyacentes
            for neighbor in nodes:
                if neighbor not in visited:  # Si no se ha visitado
                    # Calcular el costo total
                    new_visited = visited | {neighbor}
                    cost = self._get_edge_weight(current_node, neighbor) + tsp(new_visited, neighbor)

                    # Actualizamos el costo mínimo y la ruta si es mejor
                    if cost < min_cost:
                        min_cost = cost
                        path[key] = neighbor  # Guardamos la mejor siguiente opción

            memo[key] = min_cost
            return min_cost

        # Iniciar el recorrido
        optimal_cost = tsp({start_vertex}, start_vertex)

        # Reconstruir el camino óptimo
        optimal_path = [start_vertex]
        current_key = (frozenset({start_vertex}), start_vertex)

        while current_key in path:
            next_node = path[current_key]
            optimal_path.append(next_node)
            current_key = (frozenset(optimal_path), next_node)

        return optimal_cost, optimal_path
````

### Explicación del Método `tsp_small_graph`:

1. **Inicialización**:
   - Se obtienen todos los nodos adyacentes al nodo de inicio.
   - Se inicializan dos estructuras: `memo` para almacenar los costos mínimos y `path` para rastrear de dónde se obtuvo el mejor costo.
2. **Función Recursiva `tsp`**:
   - Esta función calcula el costo del TSP de manera recursiva:
     - **Condición de Base**: Si todos los nodos han sido visitados, retorna el costo de volver al nodo de inicio.
     - Se usa una clave única (una tupla del conjunto de nodos visitados y el nodo actual) para saber si ya hemos calculado este estado.
3. **Exploración de Vecinos**:
   - Para cada nodo vecino no visitado, se calcula el costo de ir a ese nodo y luego de hacer la llamada recursiva para el próximo estado con ese nodo marcado como visitado.
   - Si se encuentra un costo menor, se actualiza `min_cost` y se guarda en `path`.
4. **Reconstruction de Resultados**:
   - Después de que se calcula el costo óptimo, se reconstruye la ruta óptima utilizando la información almacenada en `path`.

### Complejidad:

- Este algoritmo tiene una complejidad de $O(n^2 \cdot 2^n)$, siendo $n$ el número de nodos. Esto es mucho más eficiente para grafos pequeños (hasta 10 nodos) que la búsqueda de todas las permutaciones.

El objetivo del TSP es encontrar la ruta más corta que visite cada nodo una vez y regrese al nodo de inicio. Utilizamos una técnica de programación dinámica que representa subconjunto
s de nodos visitados usando máscaras de bits, junto con una memorización para optimizar el cálculo.

### Código Comentado

````python
import itertools  # Importamos itertools para usar combinaciones

class GrafoAvanzado(GrafoAvanzado):
    # Otros métodos de la clase...

    def tsp_small_graph_bits(self, start_vertex):
        """
        Resuelve el Problema del Viajante de Comercio para un grafo completo pequeño (~10 nodos) comenzando desde un nodo especificado. 
        Se requiere encontrar el recorrido óptimo. Se esperan grafos con un máximo de 10 nodos. Debe ejecutarse en menos de 1 segundo.

        Parámetros:
        - `start_vertex`: El nodo de inicio.

        Retornos:
        - Una tupla que contiene la distancia total del recorrido y una lista de nodos que representan el camino del recorrido.
        """
        # Asegúrese de que el vértice inicial esté en el grafo
        if start_vertex not in self.graph:
            raise KeyError("El vértice inicial debe existir en el gráfico.")
        
        # Lista de vértices excluyendo el vértice inicial
        vertices = list(self.graph.keys())
        vertices.remove(start_vertex)  # Retiramos el nodo de inicio de la lista
        
        # Número de vértices
        n = len(vertices)
        
        # Mapeo de vértices por índice para fácil acceso
        index_map = {vertex: i for i, vertex in enumerate(vertices)}
        
        # Inicializamos la memoria (memoización) para almacenar los costos mínimos
        memo = {}
        # Caso base: distancia desde el vértice inicial a cada vértice en el conjunto
        for i in range(n):
            # Generamos claves únicas para la memoria usando desplazamiento de bits
            # Almacenamos la distancia desde el vértice inicial al vértice 'i'
            memo[(1 << i, i)] = (self._get_edge_weight(start_vertex, vertices[i]), None)
        
        # Iteramos sobre tamaños de subconjuntos de 2 a n (número total de vértices)
        for subset_size in range(2, n + 1):
            # Generamos todos los subconjuntos de un tamaño dado
            for subset in itertools.combinations(range(n), subset_size):
                # Creamos una máscara que representa el subconjunto actual
                subset_mask = sum(1 << i for i in subset)
                
                # Calculamos la ruta más corta para cada vértice final en el subconjunto
                for endpoint in subset:
                    # Eliminamos el vértice "endpoint" de la máscara para obtener el "prev_mask"
                    prev_mask = subset_mask & ~(1 << endpoint)
                    min_dist = float('inf')  # Inicializamos la distancia mínima como infinita
                    previous_vertex = None  # Variable para rastrear el vértice anterior
                    
                    # Encuentra la ruta más corta al vértice final
                    for prev in subset:
                        if prev == endpoint:  # Ignoramos el caso en que 'prev' es igual a 'endpoint'
                            continue
                        if (prev_mask, prev) in memo:  # Verificamos si el estado está en la memoria
                            dist, _ = memo[(prev_mask, prev)]
                            # Calculamos la nueva distancia al agregar el peso de la arista al 'endpoint'
                            dist += self._get_edge_weight(vertices[prev], vertices[endpoint])
                            
                            # Si encontramos una distancia menor, actualizamos min_dist y previous_vertex
                            if dist < min_dist:
                                min_dist = dist
                                previous_vertex = prev
                    
                    # Almacenamos la distancia mínima y el vértice anterior en la memoria
                    memo[(subset_mask, endpoint)] = (min_dist, previous_vertex)
        
        # Busca la ruta más corta que regresa al vértice inicial
        min_distance = float('inf')  # Inicializamos la distancia mínima como infinita
        last_vertex = None  
        full_mask = (1 << n) - 1  # Máscara que representa todos los vértices visitados
        
        for i in range(n):
            if (full_mask, i) in memo:  # Verificamos si el estado de todos los vértices está en la memoria
                dist, _ = memo[(full_mask, i)]
                # Agregamos el costo de regresar al 'start_vertex'
                dist += self._get_edge_weight(vertices[i], start_vertex)
                
                # Si encontramos una distancia menor, actualizamos min_distance y last_vertex
                if dist < min_distance:
                    min_distance = dist
                    last_vertex = i
        
        # Reconstruir el camino más corto
        path = [start_vertex]  # Comenzamos el camino desde el nodo inicial
        mask = full_mask  # Restablecemos la máscara completa
        current_vertex = last_vertex  # Comenzamos desde el último vértice
        
        while current_vertex is not None:  # Mientras haya un vértice actual
            path.append(vertices[current_vertex])  # Agrega el vértice actual al camino
            # Recupera el siguiente vértice en el camino desde la memoria
            next_vertex = memo[(mask, current_vertex)][1]
            # Actualizamos la máscara para eliminar el vértice actual
            mask &= ~(1 << current_vertex)
            # Actualiza 'current_vertex' para el siguiente paso
            current_vertex = next_vertex
        
        path.append(start_vertex)  # Agregar el nodo inicial al final del camino
        
        return min_distance, path  # Retornamos la distancia mínima y el camino
````

````python
import itertools  # Importamos itertools para usar combinaciones
from functools import lru_cache

class GrafoAvanzado(GrafoAvanzado):
    # Otros métodos de la clase...

    def tsp_small_graph_bits(self, start_vertex):
        """
        Resuelve el Problema del Viajante de Comercio para un grafo completo pequeño (~10 nodos) comenzando desde un nodo especificado. 
        Se requiere encontrar el recorrido óptimo. Se esperan grafos con un máximo de 10 nodos. Debe ejecutarse en menos de 1 segundo.

        Parámetros:
        - `start_vertex`: El nodo de inicio.

        Retornos:
        - Una tupla que contiene la distancia total del recorrido y una lista de nodos que representan el camino del recorrido.
        """
        # Asegúrese de que el vértice inicial esté en el grafo
        if start_vertex not in self.graph:
            raise KeyError("El vértice inicial debe existir en el gráfico.")
        
        # Lista de vértices excluyendo el vértice inicial
        vertices = list(self.graph.keys())
        vertices.remove(start_vertex)  # Retiramos el nodo de inicio de la lista
        
        # Número de vértices
        n = len(vertices)
        
        # Mapeo de vértices por índice para fácil acceso
        index_map = {vertex: i for i, vertex in enumerate(vertices)}
        
        # Inicializamos la memoria (memoización) para almacenar los costos mínimos
        memo = {}
        # Caso base: distancia desde el vértice inicial a cada vértice en el conjunto
        for i in range(n):
            # Generamos claves únicas para la memoria usando desplazamiento de bits
            # Almacenamos la distancia desde el vértice inicial al vértice 'i'
            memo[(1 << i, i)] = (self._get_edge_weight(start_vertex, vertices[i]), None)
        
        # Iteramos sobre tamaños de subconjuntos de 2 a n (número total de vértices)
        for subset_size in range(2, n + 1):
            # Generamos todos los subconjuntos de un tamaño dado
            for subset in itertools.combinations(range(n), subset_size):
                # Creamos una máscara que representa el subconjunto actual
                subset_mask = sum(1 << i for i in subset)
                
                # Calculamos la ruta más corta para cada vértice final en el subconjunto
                for endpoint in subset:
                    # Eliminamos el vértice "endpoint" de la máscara para obtener el "prev_mask"
                    prev_mask = subset_mask & ~(1 << endpoint)
                    min_dist = float('inf')  # Inicializamos la distancia mínima como infinita
                    previous_vertex = None  # Variable para rastrear el vértice anterior
                    
                    # Encuentra la ruta más corta al vértice final
                    for prev in subset:
                        if prev == endpoint:  # Ignoramos el caso en que 'prev' es igual a 'endpoint'
                            continue
                        if (prev_mask, prev) in memo:  # Verificamos si el estado está en la memoria
                            dist, _ = memo[(prev_mask, prev)]
                            # Calculamos la nueva distancia al agregar el peso de la arista al 'endpoint'
                            dist += self._get_edge_weight(vertices[prev], vertices[endpoint])
                            
                            # Si encontramos una distancia menor, actualizamos min_dist y previous_vertex
                            if dist < min_dist:
                                min_dist = dist
                                previous_vertex = prev
                    
                    # Almacenamos la distancia mínima y el vértice anterior en la memoria
                    memo[(subset_mask, endpoint)] = (min_dist, previous_vertex)
        
        # Busca la ruta más corta que regresa al vértice inicial
        min_distance = float('inf')  # Inicializamos la distancia mínima como infinita
        last_vertex = None  
        full_mask = (1 << n) - 1  # Máscara que representa todos los vértices visitados
        
        for i in range(n):
            if (full_mask, i) in memo:  # Verificamos si el estado de todos los vértices está en la memoria
                dist, _ = memo[(full_mask, i)]
                # Agregamos el costo de regresar al 'start_vertex'
                dist += self._get_edge_weight(vertices[i], start_vertex)
                
                # Si encontramos una distancia menor, actualizamos min_distance y last_vertex
                if dist < min_distance:
                    min_distance = dist
                    last_vertex = i
        
        # Reconstruir el camino más corto
        path = [start_vertex]  # Comenzamos el camino desde el nodo inicial
        mask = full_mask  # Restablecemos la máscara completa
        current_vertex = last_vertex  # Comenzamos desde el último vértice
        
        while current_vertex is not None:  # Mientras haya un vértice actual
            path.append(vertices[current_vertex])  # Agrega el vértice actual al camino
            # Recupera el siguiente vértice en el camino desde la memoria
            next_vertex = memo[(mask, current_vertex)][1]
            # Actualizamos la máscara para eliminar el vértice actual
            mask &= ~(1 << current_vertex)
            # Actualiza 'current_vertex' para el siguiente paso
            current_vertex = next_vertex
        
        path.append(start_vertex)  # Agregar el nodo inicial al final del camino
        
        return min_distance, path
    
    def tsp_small_graph_bits2(self, start_vertex): 
        """
        Resuelve el Problema del Viajante de Comercio (TSP) para un grafo completo pequeño (~10 nodos)
        comenzando desde un nodo específico. Debe encontrar el recorrido óptimo y ejecutarse en menos de 1 segundo.

        Parámetros:
        - start_vertex: El nodo de inicio.

        Retorna:
        - Una tupla que contiene la distancia total del recorrido y una lista de nodos que representan el camino del recorrido.
        """
        from itertools import permutations
        from functools import lru_cache

        # Validamos que el nodo inicial exista
        if start_vertex not in self.graph:
            raise KeyError(f"El nodo '{start_vertex}' no existe en el grafo")

        vertices = list(self.graph.keys())
        n = len(vertices)
        start_index = vertices.index(start_vertex)

        # Creamos una matriz de distancias para acceso O(1)
        dist_matrix = [[float('inf')] * n for _ in range(n)]
        for i, src in enumerate(vertices):
            for j, dest in enumerate(vertices):
                dist_matrix[i][j] = self._get_edge_weight(src, dest)

        @lru_cache(None)
        def tsp(mask, pos):
            # Caso base: si visitamos todos los nodos, volvemos al nodo inicial
            if mask == (1 << n) - 1:
                return dist_matrix[pos][start_index]

            ans = float('inf')
            # Probar todas las siguientes ciudades
            for next_city in range(n):
                if mask & (1 << next_city) == 0:
                    new_dist = dist_matrix[pos][next_city] + tsp(mask | (1 << next_city), next_city)
                    ans = min(ans, new_dist)
            return ans

        # Calcular la distancia mínima
        min_distance = tsp(1 << start_index, start_index)

        # Reconstruir el camino
        path = [start_vertex]
        mask = 1 << start_index
        pos = start_index
        for _ in range(n - 1):
            next_city = None
            best_dist = float('inf')
            for city in range(n):
                if mask & (1 << city) == 0:
                    dist_to_city = dist_matrix[pos][city] + tsp(mask | (1 << city), city)
                    if dist_to_city < best_dist:
                        best_dist = dist_to_city
                        next_city = city
            path.append(vertices[next_city])
            mask |= 1 << next_city
            pos = next_city

        path.append(start_vertex)
        return min_distance, path
````

````python
import time
import heapq

class GrafoAvanzado(GrafoAvanzado):

    def get_mst(self):
        """Genera un Árbol de Expansión Mínima usando el algoritmo de Prim."""
        start = next(iter(self.graph))
        visited = set()
        min_heap = [(0, start)]
        mst_edges = []

        while min_heap:
            weight, vertex = heapq.heappop(min_heap)
            if vertex in visited:
                continue
            
            visited.add(vertex)
            mst_edges.append((weight, vertex))
            
            for neighbor, edge_weight in self.graph[vertex].items():
                if neighbor not in visited:
                    heapq.heappush(min_heap, (edge_weight, neighbor))
        
        return mst_edges[1:]  # omite el peso inicial

    def tsp_medium_graph(self, start):
        start_time = time.time()

        # Paso 1: Obtener el MST
        mst_edges = self.get_mst()
        total_mst_weight = sum(weight for weight, _ in mst_edges)

        # Paso 2: Generar un recorrido desde el MST
        # Para formar el recorrido, comenzamos en el nodo de inicio
        path = []
        visited = set()
        stack = [start]

        while stack:
            current = stack.pop()
            if current not in visited:
                visited.add(current)
                path.append(current)

                # Añadir adyacentes no visitados a la pila
                for neighbor in self.get_adjacent_vertices(current):
                    if neighbor not in visited:
                        stack.append(neighbor)

            # Verificamos el tiempo transcurrido
            if time.time() - start_time > 1.5:
                break

        # Paso 3: Completar recorrido volviendo al inicio
        total_cost = 0
        for i in range(len(path) - 1):
            total_cost += self._get_edge_weight(path[i], path[i + 1])
        
        # Agregar el regreso al nodo inicial
        total_cost += self._get_edge_weight(path[-1], start)
        path.append(start)

        return total_cost, path
````

1. **MST usando Prim:** El método `get_mst` genera un Árbol de Expansión Mínima con el algoritmo de Prim. Se utiliza una cola de prioridad para mantener los vértices que se añadirán a
l MST.
2. **Construcción de Recorrido:** Utilizamos una pila para emular un recorrido en profundidad (DFS) del MST, asegurando que evitamos visitar nodos que ya fueron incluidos. Esto mejora
 la eficiencia del recorrido.

````python
from math import exp
from random import sample

class GrafoAvanzado(GrafoAvanzado):  
    def tsp_medium_graph(self, start):
        """
        Resolver el Problema del Viajante de Comercio para un grafo completo de tamaño mediano (~300 nodos) 
        comenzando desde un nodo especificado. Se espera que funcione mejor que tsp_large_graph. 
        Debe ejecutarse en menos de 1.5 segundos.
        
        Parámetros:
        start: El nodo de inicio.
        
        Devuelve:
        Una tupla que contiene la distancia total del recorrido y una lista de nodos que representan el camino del recorrido.
        """
        if start not in self.graph:
            raise KeyError("El vértice inicial debe existir en el gráfico.")
    
        # Heurística de Vecino Más Cercano para generar un recorrido inicial
        unvisited = set(self.graph.keys())  # Conjunto de nodos no visitados
        unvisited.remove(start)  # Eliminamos el nodo inicial
        current_node = start  # Primer nodo para arrancar
        tour = [start]  # Ruta con el inicio
    
        while unvisited:  # Mientras exista nodo por visitar
            # Buscamos el nodo con la menor distancia del nodo actual
            nearest_node = min(unvisited, 
                               key=lambda node: self._get_edge_weight(current_node, node))
            tour.append(nearest_node)  # Agregamos el nodo a la ruta
            current_node = nearest_node  # Cambiamos de nodo
            unvisited.remove(nearest_node)  # Eliminamos nodo visitado
    
        # Agregamos el nodo de inicio para cerrar el recorrido
        tour.append(start) 
        
        # Optimización mediante Simulated Annealing
        def calculate_total_distance(tour):
            """Calcula la distancia total del tour."""
            return sum(self._get_edge_weight(tour[i], tour[i + 1]) for i in range(len(tour) - 1))
    
        def simulated_annealing(tour, initial_temp, cooling_rate, max_iterations):
            """
            Optimiza un tour utilizando el algoritmo de Simulated Annealing.
    
            Parámetros:
                tour (list): Lista de nodos que representan el recorrido inicial.
                initial_temp (float): Temperatura inicial que controla la probabilidad de aceptar soluciones peores.
                cooling_rate (float): Tasa de enfriamiento que determina cómo disminuye la temperatura en cada iteración.
                max_iterations (int): Número máximo de iteraciones a realizar para optimizar el tour.
    
            Devuelve:
                tuple: Una tupla que contiene:
                    - list: El tour optimizado.
                    - float: La distancia total del tour optimizado.
            """
            current_distance = calculate_total_distance(tour)  # Calcula la distancia total del tour
            best_tour = tour[:]  # Copia el tour
            best_distance = current_distance  # Toma la distancia total como la mejor
            temperature = initial_temp  # Copia el valor de la temperatura 
            
            for _ in range(max_iterations):  # Itera por el número máximo de iteraciones
                # Selecciona aleatoriamente dos índices `i` y `j` del recorrido
                i, j = sorted(sample(range(1, len(tour) - 1), 2))  
                # Genera una nueva ruta al invertir la sección del tour entre i y j
                new_tour = tour[:i] + tour[i:j][::-1] + tour[j:]
                new_distance = calculate_total_distance(new_tour)
    
                # Condición de aceptación
                if (new_distance < best_distance or 
                    random.random() < exp((current_distance - new_distance) / temperature)):
                    tour = new_tour
                    current_distance = new_distance
                    if new_distance < best_distance:
                        best_tour = new_tour
                        best_distance = new_distance
                
                # Enfriamiento de la temperatura
                temperature *= cooling_rate 
            
            return best_tour, best_distance
    
        # Llamada al algoritmo de Simulated Annealing
        optimized_tour, optimized_distance = simulated_annealing(tour, initial_temp=1000, cooling_rate=0.995, max_iterations=1000)
        
        return optimized_distance, optimized_tour
````
